```python
shoppingCart=''
for fruit in range(5):
  saved=input('Type a fruit: ')
  if fruit==0:
    shoppingCart=saved
  else:
    shoppingCart=shoppingCart+', '+saved
print(shoppingCart)
```

```python
iterations=int(input('Type the amount of numbers to add: '))
sum=0
for number in range(iterations):
  numberToAdd=int(input('Type the '+str(number+1)+' number to add: '))
  sum=sum+numberToAdd
print('The sum is',sum)
```
