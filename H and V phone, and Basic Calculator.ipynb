Superhero Or Villian Phone

```python
#Variable asks the user for input and only respects the two, else it spits out else's code.
option=int(input('do you want to call a Hero or a Villian? Type 1 for Hero, Type 2 for Villian. '))
if(option==1):
  print('You are calling a Hero...')
  print('What Hero do you want to call? ')
  #\n creates page breaks, do not utilize spaces when using
  print('1. Iron Man \n2. Spider-Man \n3. Goku')
  #Code below reuses the option variable to output the associated hero to assigned integer input
  option=int(input('Please type the number related to the Hero '))
  #if elif and else demonstration to show progression in hierarchy, if a value is true it outputs assigned text
  if(option==1):
    print('You are calling Iron Man...')
  elif(option==2):
    print('You are calling Spider-Man...')
  elif(option==3):
    print('You are calling Goku...')
  else:
    print('Please type the right option')
elif(option==2):
  print('You are calling a Villian...')
  print('What Villian do you want to call? ')
  print('1. Dr. Doom \n2. Green Goblin \n3. Vegeta')
  #Reused option=int... code due to the variable not being needed after the initial input
  option=int(input('Please type the number related to the Villain '))
  if(option==1):
    print('You are calling Dr. Doom...')
  elif(option==2):
    print('You are calling Green Goblin...')
  elif(option==3):
    print('You are calling Vegeta...')
  else:
    print('Please type the right option')
else:
  print('Please type the right option')
```

Calculator

```python
print('Welcome user to your calculator :')
print('Press \n1. For Addition\n2. For Subtraction')
calcOption=int(input('Type the number related to the option '))
if(calcOption==1):
  print('You have selected Addition')
  variableA=int(input('Type the number A you want to add '))
  variableB=int(input('Type the number B you want to add '))
  #if the variable is over 0 then it will add and print the total
  if(variableA>=0 and variableB>=0):
    total=variableA+variableB
    print('Your total is',total)
  else:
    print('One or both numbers are negative')
elif(calcOption==2):
  print('You have selected Subtraction')
  variableA=int(input('Type the number A you want to subtract '))
  variableB=int(input('Type the number B you wnat to subtract '))
  if(variableA>=0 and variableB>=0):
    #Makes sure Variable A is Greater than or Equal to Variable B else it prints a negative
    if(variableA>=variableB):
      total=variableA-variableB
      print('Your total is',total)
    else:
      print('your total will be negative')
  else:
    print('One or both of the numbers are negative')
else:
  print('Please select a valid option')
```
