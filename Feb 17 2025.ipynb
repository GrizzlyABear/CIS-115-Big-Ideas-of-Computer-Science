```python
year=int(input('Please insert or type a year to evaluate format yyyy: '))
if(year<=0):
  print('Please input a valid year')
else:
  if(year%4==0):
    if(year%100==0):
      if(year%400==0):
        print('its a leap year ')
      else:
        print('it is not a leap year ')
    else:
      print('it is a leap year ')
  else:
    print('it is not a leap year ')
```

Range A: 90-100
Range B: 80-89
Range C: 70-79
Range D: 60-69
Range F: 0-59

```python
grade=int(input('Please type a number from 0-100 to make the grading calculations: '))
if(grade>89 and grade<=100): #A
  print('Grade is an A')
elif(grade>=80 and grade<90): #B
  print('Grade is a B')
elif(grade>=70 and grade<=79): #C
  print('Grade is a C')
elif(grade>=60 and grade<=69): #D
  print('Grade is a D')
elif(grade>=0 and grade<60): #F
  print('Grade is an F')
else:
  print('Please input a valid number')
```
